```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
data = pd.read_csv('Advertising.csv')
data
```

(a)

```python
X = data.drop('Sales',axis=1)
y = data['Sales']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
print('Intercept: ',model.intercept_)
print('Coefficient: ', model.coef_)
```

```python
pred = model.predict(X_test)
```

```python
print('Actual Values: ', y_test.head(5))
print('Predicted Values: ', pred[:5])
```

```python
plt.scatter(X_test.values.ravel(), y_test, color='blue', label='Actual Sales')  # flatten X_test
plt.plot(X_test.values.ravel(), pred, color='red', linewidth=2, label='Regression Line')
plt.xlabel('TV Advertising Budget')
plt.ylabel('Sales')
plt.title('Linear Regression: TV vs Sales')
plt.legend()
plt.show()
```

```python
r2_score(y_test, pred)
```

```python
mean_absolute_error(y_test, pred)
```

```python
mse = mean_squared_error(y_test, pred)
mse
```

```python
rmse = np.sqrt(mse)
rmse
```
